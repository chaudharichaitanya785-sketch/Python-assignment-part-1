```python
# Raw data
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

for student in raw_students:
    # Clean name
    name = student["name"].strip().title()

    # Convert roll
    roll = int(student["roll"])

    # Convert marks
    marks = [int(x) for x in student["marks_str"].split(", ")]

    # Name validation
    is_valid = all(word.isalpha() for word in name.split())

    if is_valid:
        print(f"{name} → ✓ Valid name")
    else:
        print(f"{name} → ✗ Invalid name")

    # Store cleaned data
    cleaned_students.append({
        "name": name,
        "roll": roll,
        "marks": marks
    })

    # Profile card
    print("="*32)
    print(f"Student : {name}")
    print(f"Roll No : {roll}")
    print(f"Marks   : {marks}")
    print("="*32)

# Find roll 103
for student in cleaned_students:
    if student["roll"] == 103:
        print("\nSpecial Output:")
        print(student["name"].upper())
        print(student["name"].lower())
```

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

print("\nSubject-wise Grades:")

for i in range(len(subjects)):
    mark = marks[i]

    if mark >= 90:
        grade = "A+"
    elif mark >= 80:
        grade = "A"
    elif mark >= 70:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "F"

    print(f"{subjects[i]} : {mark} → {grade}")

# Calculations
total = sum(marks)
average = round(total / len(marks), 2)

print("\nTotal Marks:", total)
print("Average:", average)

# Highest & Lowest
max_mark = max(marks)
min_mark = min(marks)

print("Highest:", subjects[marks.index(max_mark)], max_mark)
print("Lowest :", subjects[marks.index(min_mark)], min_mark)

# While loop for new subjects
new_subjects = []
new_marks = []

while True:
    sub = input("Enter subject name (or 'done'): ")

    if sub.lower() == "done":
        break

    mark_input = input("Enter marks (0-100): ")

    if not mark_input.isdigit():
        print("Invalid input! Must be a number.")
        continue

    mark = int(mark_input)

    if mark < 0 or mark > 100:
        print("Invalid marks! Must be between 0-100.")
        continue

    new_subjects.append(sub)
    new_marks.append(mark)

print("\nNew subjects added:", len(new_subjects))

# Updated average
all_marks = marks + new_marks
updated_avg = round(sum(all_marks) / len(all_marks), 2)

print("Updated Average:", updated_avg)
```

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

print("\nClass Report")
print("Name              | Average | Status")
print("-"*40)

pass_count = 0
fail_count = 0
averages = []

topper_name = ""
topper_avg = 0

for name, marks in class_data:
    avg = round(sum(marks) / len(marks), 2)
    averages.append(avg)

    status = "Pass" if avg >= 60 else "Fail"

    if status == "Pass":
        pass_count += 1
    else:
        fail_count += 1

    print(f"{name:<18} | {avg:<7} | {status}")

    # Topper logic
    if avg > topper_avg:
        topper_avg = avg
        topper_name = name

# Final stats
class_avg = round(sum(averages) / len(averages), 2)

print("\nPassed:", pass_count)
print("Failed:", fail_count)
print("Topper:", topper_name, topper_avg)
print("Class Average:", class_avg)
```
